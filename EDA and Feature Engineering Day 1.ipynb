```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
#combining both the table on countrycode colum using merge function
pd.merge(df,df_country,on = "Country Code", how = "left")
```

```python
final_df = pd.merge(df,df_country,on = "Country Code", how = "left")
final_df.head()
```

```python
final_df.columns
```

```python
country_name = final_df.Country.value_counts().index#with the respect to the countries how many records are there.
```

```python
country_value = final_df.Country.value_counts().values
```

```python
#pie chart - top 3 countries using zomato
plt.pie(country_value[:3],labels = country_name[:3], autopct="%1.2f%%")
```

Observation: Zomato has a maximum number of transaction in India.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,6) #resize the parameter
sns.barplot(x='Aggregate rating', y='Rating Count', data = ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue= 'Rating color',data = ratings, palette =['Blue','Red','Orange','Yellow','Green','Green'])
```
